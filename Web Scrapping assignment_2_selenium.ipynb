Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.

2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
driver=webdriver.Chrome(r'C:\Users\shyamanthulaRamakris\Downloads\chromedriver.exe')
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.azquotes.com/")
```

```python
Top_quotes=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
Top_quotes.click()
```

```python
Quote=[]
Author=[]
Typeof_quote=[]
start=0
end=10
for page in range(start,end):

    quote = driver.find_elements(By.XPATH,'//a[@class="title"]')
    author = driver.find_elements(By.XPATH,'//div[@class="author"]')
    typeofquote = driver.find_elements(By.XPATH,'//div[@class="tags"]')

    for i in quote:Quote.append(i.text)
    for i in author:Author.append(i.text)    
    for i in typeofquote:Typeof_quote.append(i.text)


nxt_button = driver.find_element(By.XPATH,'/html/body/div[1]/div[2]/div/div/div/div[1]/div/div[3]/li[12]/a')
nxt_button.click()
```

```python
df= pd.DataFrame({'Quote':Quote,'Author':Author,'Type of quote':Typeof_quote})
df
```
