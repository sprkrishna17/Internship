1.python program to display all the header tags from wikipedia.org and make data frame.

```python
from bs4 import BeautifulSoup
import requests
page= requests.get("https://en.wikipedia.org/wiki/Main_Page")
soup=BeautifulSoup(page.content,"html.parser")
soup
header_texts = []
for i in soup.find_all('span',class_="mw-headline"):
    header_texts.append(i.text)
header_texts
import pandas as pd
df = pd.DataFrame(header_texts, columns=["Header"])
df
```

2.python program to display list of respected former presidents of India (i.e. Name, Term ofoffice) from https://presidentofindia.nic.in/former-presidents.htm and make data frame

```python
import requests
page= requests.get('https://presidentofindia.nic.in/former-presidents.htm')
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.text, 'html.parser')
soup
results = soup.find_all('div', class_="presidentListing")
results
first_result = results[0]
first_result
first_result.find('h3').text
first_result.find('p').text
records = []
for result in results:
    PresidentName = result.find('h3').text
    TermofService = result.find('p').text
    records.append((PresidentName, TermofService))

import pandas as pd
df = pd.DataFrame(records, columns=['PresidentName', 'TermofService'])
df
```

3) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape and make data frame.

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
soup=BeautifulSoup(page.content,'html.parser')
soup
table = soup.find('table', class_='table')
teams = []
matches = []
points = []
ratings = []

for row in table.tbody.find_all('tr'):
    team = row.find('span', class_='u-hide-phablet').text.strip()
    match = row.find_all('td')[2].text.strip()
    point = row.find_all('td')[3].text.strip()
    rating = row.find_all('td')[4].text.strip()
    teams.append(team)
    matches.append(match)
    points.append(point)
    ratings.append(rating)
data = {'Team': teams, 'Matches': matches, 'Points': points, 'Rating': ratings}
df = pd.DataFrame(data)
df.index += 1  
df = df.head(10)
df
```

b) Top 10 ODI Batsmen along with the records of their team and rating.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

req=requests.get('https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
soup=BeautifulSoup(req.content)
player=soup.find_all('tr',class_=('rankings-block__banner','table-body'))
top10=player[0:10]
data={'Player_Name':[],'Team_Name': [], 'Rating':[]}
for i in top10:
    bat=i.find_all('td',recursive=True)
    data['Player_Name'].append(bat[1].text.replace('\n',''))
    data['Team_Name'].append(bat[2].text.replace('\n',''))
    data['Rating'].append(bat[3].text.replace('\n',''))
df=pd.DataFrame(data,index=range(1,11))
df
```

5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame.i) Headline ii) Time iii) News Link

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

page=requests.get('https://www.cnbc.com/world/?region=world')
page
soup=BeautifulSoup(page.content,'html.parser')
soup
Headlines=[]
time=[]
newslink=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    Headlines.append(i.text)
Headlines
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    time.append(i.text)
time
for i in soup.find_all('a',class_="LatestNews-headline"):
    newslink.append(i.get('href'))
df = pd.DataFrame ({'Headlines': Headlines,'time':time,'Newslink':newslink})
df
```

6) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles Scrape below mentioned details and make data framei) Paper Title ii) Authors iii) Published Date iv) Paper URL

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

page=requests.get('https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles')
page
soup=BeautifulSoup(page.content,'html.parser')
soup
titles = []
authors = []
dates = []
urls = []
for i in soup.find_all('h2',class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    titles.append(i.text)
titles
for i in soup.find_all('span',class_="sc-1w3fpd7-0 dnCnAO"):
    authors.append(i.text)
authors
for i in soup.find_all('span',class_="sc-1thf9ly-2 dvggWt"):
    dates.append(i.text)
dates
for i in soup.find_all('a',class_="sc-5smygv-0 fIXTHm"):
    urls.append(i.get('href'))
urls
df = pd.DataFrame({'Paper Title': titles, 'Author': authors, 'Published Date': dates, 'Paper URL': urls})
df
```

7) Write a python program to scrape mentioned details from dineout.co.inand make data frame i) Restaurant name ii) Cuisine iii) Location iv) Ratings v) Image URL

```python

import requests
from bs4 import BeautifulSoup
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
soup=BeautifulSoup(page.content)
soup

RestaurantName=soup.find('div',class_="restnt-info cursor")
RestaurantName
RestaurantName.text
RestaurantName=[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    RestaurantName.append(i.text)
RestaurantName

price=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
price

location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
location

images=[]
for i in soup.find_all('img',class_="no-img"):
    images.append(i['data-src'])
images
Ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
    
Ratings
import pandas as pd
df=pd.DataFrame({"Restaurant Name":RestaurantName, "Cusines":price, "Location":location, "Ratings":Ratings,"Images_Url":images})
df
```
